```python
import numpy as np
import pandas as pd
import re
from pathlib import Path
from skimage import io
from tqdm import tqdm
import matplotlib.pyplot as plt
```

```python
folder = Path('training_models_unet')
files = folder.glob('*/*.npy')
files = sorted([x.as_posix() for x in files])
test_files = [x for x in files if 'test' in x]
val_files = [x for x in files if 'test' not in x]
print(val_files)
```

```python
for file in val_files:
    metrics = np.load(file)
    plt.plot(metrics[:,12])
```

```python
for file in test_files:
    metrics = np.load(file)
    plt.plot(metrics[:,12])
```

```python
for file in test_files:
    metrics = np.load(file)
    print(np.mean(metrics[-200:,12]),np.std(metrics[-200:,12]),metrics[:,12][np.argmax(metrics[:,12])])
    plt.plot(metrics[:,12])
```
